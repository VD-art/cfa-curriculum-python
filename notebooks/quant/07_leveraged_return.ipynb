# 📍 Lesson 1 – Leveraged Return

- Leverage factor = total investment / own equity
- Leveraged return = unleveraged return × leverage factor

```python
from scripts.quant.leveraged_return import leveraged_return
```

```python
unleveraged = 0.10  # 10% return
leverage_factor = 1.5  # e.g., total investment is 150% of own equity

lever = leveraged_return(unleveraged, leverage_factor)
print(f"Leveraged return: {lever:.2%}")
```

```python
import matplotlib.pyplot as plt

factors = [1, 1.2, 1.5, 2, 2.5]
leveraged_returns = [leveraged_return(unleveraged, f) for f in factors]

plt.plot(factors, leveraged_returns, marker='o')
plt.xlabel('Leverage factor')
plt.ylabel('Leveraged return')
plt.title('Effect of leverage on return')
plt.grid()
plt.show()
```
