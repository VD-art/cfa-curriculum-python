We break required return into:
- Real risk-free rate
- Inflation premium
- Default risk premium
- Liquidity premium
- Maturity premium

Required return = sum of these components.

```python
from scripts.quant.required_return_components import calculate_required_return_components
```

```python
inputs = {
    'real_risk_free_rate': 0.02,
    'inflation_premium': 0.015,
    'default_risk_premium': 0.01,
    'liquidity_premium': 0.005,
    'maturity_premium': 0.004
}

result = calculate_required_return_components(**inputs)

print("Components:")
for k, v in result.items():
    print(f"{k}: {v:.2%}")
```

```python
import matplotlib.pyplot as plt

components = list(result.keys())[:-1]
values = [result[c] for c in components]

plt.barh(components, values)
plt.xlabel('Contribution')
plt.title('Required Return Components')
plt.show()
```
